```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from PIL import Image

model_path = 'runs/detect/train/weights/best.pt'
model = YOLO(model_path)

img_paths = [
    '/content/extracted_files/DS-71c1fd51-v2/images/val/aiformankind_v2_001859.jpg',
    '/content/extracted_files/DS-71c1fd51-v2/images/val/aiformankind_v2_001762.jpg',
    '/content/extracted_files/DS-71c1fd51-v2/images/val/aiformankind_v1_000706.jpg',
    '/content/extracted_files/DS-71c1fd51-v2/images/val/ADF_1320_2023_05_23T18_27_40.jpg'
]

plt.figure(figsize=(12, 12))

for i, img_path in enumerate(img_paths):
    image = Image.open(img_path).convert("RGB")
    input_tensor = np.array(image)

    results = model(input_tensor)

    predictions = results[0].boxes
    boxes = predictions.xyxy.cpu().numpy()
    scores = predictions.conf.cpu().numpy()
    classes = predictions.cls.cpu().numpy()

    original_image = cv2.imread(img_path)
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    for box, score, cls in zip(boxes, scores, classes):
        x1, y1, x2, y2 = box.astype(int)
        cv2.rectangle(original_image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Boîte en rouge
        cv2.putText(original_image, f'Class: {int(cls)}, Score: {score:.2f}', (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    plt.subplot(2, 2, i + 1)
    plt.imshow(original_image)
    plt.axis('off')
    plt.title(f'Image {i + 1}')

plt.tight_layout()
plt.show()
```
